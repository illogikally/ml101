```python
import numpy as np

data = np.loadtxt("nn.txt", delimiter=",")
X = data[:, :-1]
y = data[:, -1]

np.random.seed(0)
w = np.random.random_sample(y.shape)
b = np.random.random()
```

```python
def a(z):
    if z > 0: return 1
    else: return -1
        
    

for loop in range(100):
    for i in range(X[0].shape[0]):
        e = a(np.dot(X[:, i], w) + b) - y[i]
        print(e)
        w += e*X[:, i]
        b += b + e*b
 
print(w)
print(b)
        
    
    
```
